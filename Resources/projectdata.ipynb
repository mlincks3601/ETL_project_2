```python
#create our path to our csv files
import os
import numpy
import pandas as pd
import sqlalchemy 
from sqlalchemy import create_engine
import psycopg2
```

```python
#read our 1st csv file
uni_rankings_path = os.path.join("../Resources/university-rankings-results.csv")
uni_rankings_df = pd.read_csv(uni_rankings_path, encoding='latin-1')
uni_rankings_df.head()
```

```python
#read our 2nd csv file
salaries_region_path = os.path.join("../Resources/salaries-by-region.csv")
salaries_region_df = pd.read_csv(salaries_region_path, encoding='latin-1')
salaries_region_df.head()
```

```python
#read our 3rd csv file
salaries_college_type_path = os.path.join("../Resources/salaries-by-college-type.csv")
salaries_college_type_df = pd.read_csv(salaries_college_type_path, encoding='latin-1')
salaries_college_type_df.head()
```

```python
#create a new dataframe with columns we do want for the 1st df
new_uni_rankings_df = uni_rankings_df[['name', 'rank', 'state']]
new_uni_rankings_df.set_index("name", inplace = True)
new_uni_rankings_df.head()

#Save dataframe to a csv file for further data exploration if needed
#new_uni_rankings_df.to_csv('C:/Users/mlinc/Desktop/new_uni_rankings_df.csv')
```

```python
#create a new dataframe with columns we do want for the 2nd df
new_salaries_region_df = salaries_region_df[['School Name', 'Starting Median Salary']]
new_salaries_region_df.set_index("School Name", inplace = True)
new_salaries_region_df.head()

#Save dataframe to a csv file for further data exploration if needed
#new_salaries_region_df.to_csv('C:/Users/mlinc/Desktop/new_salaries_region_df.csv')
```

```python
#merge our two new data tables together
merged_rank_and_salary_df = pd.merge(new_salaries_region_df, new_uni_rankings_df, left_index=True, right_index=True)
merged_rank_and_salary_df.head(10)
```

```python
#sort the rank by ascending order
ascnd_by_rank_merged_df = merged_rank_and_salary_df.sort_values(by="rank",
                                       ascending=True)
ascnd_by_rank_merged_df.head(10)
```

```python
#sort the salary by descending order to see the highest earning salary University first
dscnd_by_salary_merged_df = merged_rank_and_salary_df.sort_values(by="Starting Median Salary",
                                       ascending=False)
dscnd_by_salary_merged_df.head(10)
```
